```python
import cianparser
import pandas as pd
import sqlite3
```

<h1>СПИСОК ГОРОДОВ</h1>

```python
locations = cianparser.list_locations()
# locations
```

<h1>НАСТРОЙКИ</h1>

```python
location_parser = cianparser.CianParser(location="Москва")
deal_type = "sale"
with_saving_csv = True
with_extra_data = True
additional_settings = {
    "start_page":1, 
    "end_page":10
}
```

<h1>ПРЕДЛОЖЕНИЯ ПО ПРОДАЖЕ КВАРТИР</h1>

```python
flats = location_parser.get_flats(
    deal_type=deal_type, 
    rooms="all", 
    with_saving_csv=with_saving_csv,
    with_extra_data=with_extra_data,
    additional_settings=additional_settings
)

print(flats[0])
```

```python
flats_df = pd.json_normalize(flats)
```

```python
flats_df.head()
```

```python
flats_df.shape
```

<h1>ПРЕДЛОЖЕНИЯ ПО ПРОДАЖЕ ДОМОВ</h1>

```python
houses = location_parser.get_suburban(
    suburban_type="house", 
    deal_type=deal_type, 
    with_saving_csv=with_saving_csv,
    with_extra_data=with_extra_data,
    additional_settings=additional_settings
)

print(houses[0])
```

```python
houses_df = pd.json_normalize(houses)
```

```python
houses_df.head()
```

```python
houses_df.shape
```

<h1>ПРЕДЛОЖЕНИЯ ПО ПРОДАЖЕ ТАУНХАУСОВ</h1>

```python
townhouses = location_parser.get_suburban(
    suburban_type="townhouse", 
    deal_type=deal_type, 
    with_saving_csv=with_saving_csv,
    with_extra_data=with_extra_data,
    additional_settings=additional_settings
)

print(townhouses[0])
```

```python
townhouses_df = pd.json_normalize(townhouses)
```

```python
townhouses_df.head()
```

```python
townhouses_df.shape
```

<h1>СОЕДИНЯЕМ ВСЕ ВМЕСТЕ</h1>

```python
real_estate = list()
real_estate.extend(flats)
real_estate.extend(houses)
real_estate.extend(townhouses)
```

```python
real_estate_df = pd.json_normalize(real_estate)
```

```python
real_estate_df.head()
```

```python
real_estate_df.shape
```

<h1>СОХРАНЯЕМ В ФАЙЛЫ</h1>

```python
# Сохраняем в csv файл
real_estate_df.to_csv('real_estate.csv', index=False)
# Сохраняем в эксель-файл
real_estate_df.to_excel('real_estate.xlsx', index=False)
# Сохраняем в файл в формате pickle
real_estate_df.to_pickle('real_estate.pkl')
```

<h1>СОХРАНЯЕМ В БАЗУ ДАННЫХ</h1>

```python
with sqlite3.connect('real_estate_db') as conn:
    real_estate_df.to_sql('real_estate', conn, if_exists='replace', index = False)
    conn.commit()
```
